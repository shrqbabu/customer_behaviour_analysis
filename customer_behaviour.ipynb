### Data Cleaning

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.groupby('Category')['Item Purchased'].unique()
```

```python
#Remapping Category to item purchased
correct_mapping = {
    #Clothing
    "T-shirt": "Clothing", "Shirt": "Clothing",
    "Shorts": "Clothing", "Hoodie": "Clothing",
    "Pants": "Clothing", "Socks": "Clothing",
    "Jeans": "Clothing", "Blouse": "Clothing",
    "Skirt": "Clothing", "Sweater": "Clothing",
    "Dress": "Clothing",

    #Accessories
    "Sunglasses": "Accessories", "Gloves": "Accessories",
    "Jewelry": "Accessories", "Hat": "Accessories",
    "Handbag": "Accessories", "Belt": "Accessories",
    "Scarf": "Accessories", "Backpack": "Accessories",
    "Bag": "Accessories",

    #Electronics
    "Laptop": "Electronics", "Phone": "Electronics",
    "Watch": "Electronics", "Headphones": "Electronics",

   #Footwear
   "Shoes": "Footwear", "Sandals": "Footwear",
   "Sneakers": "Footwear", "Boots": "Footwear",

   #Outerwear
   "Coat": "Outerwear", "Jacket": "Outerwear"
}
```

```python
df.groupby('Category')['Item Purchased'].unique()
```

```python
df['Category'] = df['Item Purchased'].map(correct_mapping)
```

```python
df.isnull().sum()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#Treating null value replacing in the Accessories category
df.loc[df['Category']== 'Accessories', 'Size'] = 'Free Size'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#Treating null value replacing in the Electronics category
df.loc[df['Category']== 'Electronics', 'Size'] = 'Not Aplicable'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#Treating null value replacing in the Footwear category
df.loc[df['Category']== 'Footwear', 'Size'] = 'Not Available'
```

```python
df.groupby('Category')['Size'].unique()
```

```python
#Treating null value replacing in the Clothing category
df.loc[(df['Category']=='Clothing') & (df['Size'].isnull()),'Size'] = df[df['Category']=='Clothing']['Size'].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.isnull().sum()
```

### Review Rating

```python
df['Review Rating'].mean()
```

```python
df['Review Rating'].median()
```

```python
df['Review Rating'].mode()
```

```python
#Treating null value replacing to the means  values
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df.isnull().sum()
```

### Previous Purchases

```python
#Treating null value Previous Purchases column replacing to the zero
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

### Purchase Amount (USD)

```python
df.groupby('Item Purchased')['Purchase Amount (USD)'].mean()
```

```python
#Treating null value Purchase Amount (USD) column replacing to the mean
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
df.isnull().sum()
```
